Import the basic files for Data_visualization

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

Load the cleaned dataset for sales


```python
sales_dv = pd.read_csv("cleaned_sales_data.csv")
```

```python
sales_dv.head()
```

To show total sales by product 

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(8, 5))

bars = ax.bar(x=sales_dv['Product'], height=sales_dv['Total_Price'], color='#1f77b4', edgecolor='black')

ax.set_title('Sales by Product', fontsize=18, fontweight='bold', color='navy')
ax.set_xlabel('Product', fontsize=14, labelpad=10)
ax.set_ylabel('Sales (in currency)', fontsize=14, labelpad=10)


plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:,.0f}',  
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5), 
                textcoords='offset points',
                ha='center', va='bottom', fontsize=10, color='black')
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sales_dv['Month'] = pd.to_datetime(sales_dv['Order_Date']).dt.month

```

```python
sales_dv['Month']
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x = sales_dv['Month'],y=sales_dv['Total_Price'],color='red')
plt.title('Monthly Sales Trend', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Month', fontsize=14, labelpad=10)
plt.ylabel('Total Sales', fontsize=14, labelpad=10)
plt.xticks(fontsize=12)
plt.show()
```

```python
plt.Figure(figsize=(8,5))
sns.boxplot(x=sales_dv['Customer_Age_Group'], y = sales_dv['Total_Price'], palette='Set2')
plt.title('Sales Distribution by Customer Age Group', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Customer Age Group', fontsize=14, labelpad=10)
plt.ylabel('Total Sales', fontsize=14, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

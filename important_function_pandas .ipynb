# Data Cleaning

 df.dropna() - Drop missing values

```python
d = {
    "one":pd.Series([12,3,7,8,np.nan]),
    "two":pd.Series([78,96,np.nan,100,120])
}
```

```python
d
```

```python
df = pd.DataFrame(d)
df
```

```python
df.dropna()
```

 df.fillna(value) - Fill missing values

```python
d = {
    "one":pd.Series([12,3,7,8,np.nan]),
    "two":pd.Series([78,96,np.nan,100,120])
} 
```

```python
df = pd.DataFrame(d)
df
```

```python
df.fillna(5 , inplace = True)
```

 df.drop(columns=['col']) - Drop column(s)

```python
df.drop(columns=['two'])
```

```python
d = {
    "one":pd.Series([12,12,7,7,7]),
    "two":pd.Series([78,96,100,100,100])
}
```

```python
df2 = pd.DataFrame(d)
```

```python
df2
```

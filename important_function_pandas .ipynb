# Data Transformation

df.apply() - Apply function to each row/column

```python
data1 = pd.read_csv("project1_df.csv")
data1
```

```python
data1.drop(columns=['Age Group'] , inplace= True)
```

```python
data1
```

```python
data1['Purchase Method'].unique()
```

```python
def purchase_method(card):
    if(card=='Credit Card'):
        return 1 
    if(card=='Debit Card'):
        return 2 
    if(card=='PhonePe UPI'):
        return 3 
    if(card == 'Google Pay UPI'):
        return 4 
    if(card == 'Net Banking'):
        return 5
    if(card=='Cash on Delivery'):
        return 6 
    if(card=='Paytm UPI'):
        return 7 
    if(card=='International Card'):
        return 8
    
```

```python
data1['Purchase_method_numaric'] = data1['Purchase Method'].apply(purchase_method)
```

```python
data1
```

```python
data1['Gender'].unique()
```

```python
def gender_group(g):
    if(g=='Male'):
        return 1
    if(g=='Female'):
        return 2
    elif(g=='Other'):
        return 3
    
```

```python
data1['Gender_numeric_format'] = data1['Gender'].apply(gender_group)
```

```python
data1
```

 df.map() - Map values in a Series

```python
data1
```

```python
data1['Discount Availed'].unique()
```

```python
def discount_numeric(dis):
    if(dis=="Yes"):
        return 1
    elif(dis=='No'):
        return 0
```

```python
data1['Discount_numeric_format'] =  data1['Discount Availed'].map(discount_numeric)
```

```python
data1
```

```python
data1.replace("Electronics","Electric_product", inplace = True)
```

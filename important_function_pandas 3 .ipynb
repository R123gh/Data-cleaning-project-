```python
import pandas as pd   
import numpy as np  
```

 Visualization (with Pandas built-in)

```python
d = {
    'X': pd.Series([1.6,78,2.3,9.3,6.3]),
    'Y':pd.Series([1,8.5,3.0,6,10])
}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

. Line Plots using Pandas DataFrame

```python
df.plot(title="Simple Line Plot using pandas", xlabel='X', ylabel='Y', figsize=(5,3))
```

Area Plots using Pandas DataFrame

```python
df.plot.area(alpha = 0.4,title="Simple  Plot area  using pandas", xlabel='X', ylabel='Y', figsize=(5,3))
```

Bar Plots using Pandas DataFrame

```python
df.plot.bar(title="Simple Bar plot using pandas", xlabel='X', ylabel='Y', figsize=(5,3))
```

Histogram Plot using Pandas DataFrame

```python
df.plot.hist(bins=50, alpha=0.5 , title="Simple Histogram using pandas", xlabel='X', ylabel='Y', figsize=(5,3))
```

Box Plots using Pandas DataFrame

```python
df.plot.box(title="Simple Box plot  using pandas", xlabel='X', ylabel='Y', figsize=(5,3))
```
